```python
from bs4 import BeautifulSoup
import requests
import csv
```

```python
file = open('labels.csv','w')
csv_writer = csv.writer(file)
csv_writer.writerow(['image name','image price'])

for i in range(0,1903):
    address = 'https://www.artprice.com/marketplace?idcurrencyzone=2&p='+str(i)+'&sort=sort_dt-desc'
    source = requests.get(address).text
    i = 0
    try:
        soup = BeautifulSoup(source, 'lxml')
        img_url = soup.find_all('div', class_='img')
        for url in img_url:
            url = url.img['src']
            print(url)
            img_splt = url.split('/')
            img_name = img_splt[-1]
            print(img_name)
            img = requests.get(url)
            open('images/'+img_name, 'wb').write(img.content)

            img_price = soup.find_all('div', class_='price')
            price = img_price[i]
            price = price.text
            i+=1
            price = price.replace(',','')
            price = str(price).strip()
            print(str(price).strip())
            
            csv_writer.writerow([img_name,price])
            print('')
    except Exception as e:
        print(e)

    
file.close()
```
